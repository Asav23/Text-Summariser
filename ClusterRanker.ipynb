```python
from collections import defaultdict

class ClusterRanker:
    def rank_clusters(
        self,
        ideas,
        idea_embeddings,
        labels,
        document_embedding
    ):
        clusters = defaultdict(list)

        for idea, emb, label in zip(ideas, idea_embeddings, labels):
            if label != -1:  # ignore noise
                clusters[label].append((idea, emb))

        ranked = []

        for label, items in clusters.items():
            size_score = len(items)

            similarity_score = np.mean([
                np.dot(emb, document_embedding) for _, emb in items
            ])

            final_score = size_score + similarity_score
            ranked.append((final_score, label, items))

        ranked.sort(reverse=True, key=lambda x: x[0])
        return ranked
```
